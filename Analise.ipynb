```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
dataframes = []
for filename in sorted(os.listdir("times/")):
    print(filename)
    dataframes.append(pd.read_csv(f"times/{filename}"))
```

```python
len(dataframes)
```

```python
columns = [column for column in dataframes[0].columns if column != "Task"]
columns
```

```python
columns = [column for column in dataframes[0].columns if column != "Task"]
rows_list = []
for row in range(len(dataframes[0])):
    frameworks_list = []
    for column in columns:
        column_list = []
        for dataframe in dataframes:
            column_list.append(dataframe[column][row])
        frameworks_list.append(column_list)
    rows_list.append(frameworks_list)  
```

```python
rows_list[-1][0]
```

```python
row = 0
for col in range(len(rows_list[row])):
    # print(columns[col])
    plt.scatter(y, rows_list[row][col])
    plt.plot(y, rows_list[row][col], label=columns[col])
    plt.title(dataframes[0]["Task"][row])
    plt.legend(loc="upper left")
    #plt.xlabel('Library')
    #plt.ylabel('Time')
    #plt.xticks(rotation=45)
    #plt.legend(title='Task')
plt.show()
```
